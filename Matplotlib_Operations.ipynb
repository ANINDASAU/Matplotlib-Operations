```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Line Plot
x = np.arange(1,11)
y = 2*x

plt.plot(x, y, color='red', linestyle=':', linewidth=3)
plt.title("Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
# Multiple Lines on Same Plot
y1 = 2*x
y2 = 3*x

plt.plot(x, y1, color='g', linestyle=':', linewidth=2, label="y=2x")
plt.plot(x, y2, color='r', linestyle='-', linewidth=3, label="y=3x")
plt.title("Multiple Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.grid(True)
plt.show()
```

```python
## Subplots
# 1 row, 2 cols, first plot
plt.subplot(1,2,1)
plt.plot(x, y1, color='g', linestyle=':', linewidth=2)
plt.title("Plot - 1")

# 1 row, 2 cols, second plot
plt.subplot(1,2,2)
plt.plot(x, y2, color='b', linestyle='-', linewidth=2)
plt.title("Plot - 2")
# prevents overlapping
plt.tight_layout()
plt.show()
```

```python
# Bar Plot
student = {'Bob':87,'Math':66,'Sam':27}
names = list(student.keys())
marks = list(student.values())

plt.bar(names, marks, color="purple")
plt.title("Bar Plot")
plt.xlabel("Names")
plt.ylabel("Marks")
plt.show()
```

```python
# Horizontal Bar Plot
plt.barh(names, marks, color="orange")
plt.title("Horizontal Bar Plot")
plt.xlabel("Marks")
plt.ylabel("Names")
plt.show()
```

```python
# Scatter Plot
x = [5,7,8,7,6,9,5,6,7,8]
y = [99,86,87,88,100,86,103,87,94,78]

plt.scatter(x, y, color="red", marker="o")
plt.title("Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
# Histogram
data = np.random.randn(1000)
plt.hist(data, bins=20, color="skyblue", edgecolor="black")
plt.title("Histogram")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```

```python
# Pie Chart
sizes = [20, 30, 25, 25]
labels = ["A","B","C","D"]
colors = ["gold","lightcoral","lightgreen","lightskyblue"]

plt.pie(sizes, labels=labels, colors=colors, autopct="%1.1f%%", startangle=90)
plt.title("Pie Chart")
plt.show()
```

```python
# Stack Plot
days = [1,2,3,4,5]
eating = [2,3,4,3,2]
sleeping = [7,8,6,11,7]
working = [7,8,7,2,2]
playing = [8,5,7,8,13]

plt.stackplot(days, eating, sleeping, working, playing,
              labels=["Eating","Sleeping","Working","Playing"],
              colors=["m","c","r","k"])
plt.legend(loc="upper left")
plt.title("Stack Plot")
plt.show()
```

```python
# Fill Between (Area Plot)
x = np.linspace(0,10,100)
y = np.sin(x)

plt.plot(x, y, color="blue")
plt.fill_between(x, y, alpha=0.3)
plt.title("Area Plot (Fill Between)")
plt.show()
```
